```python
def add_not_present_strings(input_list):
    result_list = []

    for item in input_list:
        parts = item.split('_', 1)
        prefix = parts[0]
        suffix = parts[1]



        sub_not_present_suffix = "count_subject_not_present"
        rel_not_present_suffix = "count_relative_not_present"
        #print(prefix, """--""",suffix)

        if "count_subject_present" in suffix:
            print("hit ", prefix, suffix)
            new_suffix = suffix.replace("count_subject_present", sub_not_present_suffix)
            result_list.append(f"{prefix}_{new_suffix}")

        if "count_relative_present" in suffix:
            new_suffix = suffix.replace("count_relative_present", rel_not_present_suffix)
            result_list.append(f"{prefix}_{new_suffix}")

    return result_list

# Test the function with your input list


result = add_not_present_strings(strings)
for item in result:
    print(item)
```
